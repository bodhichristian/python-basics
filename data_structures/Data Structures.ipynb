# Data Structures

## Classification

### Linear Data Structure

Data elements are arranged sequentially or linearly, where each element is attached to its previous and next adjacent elements.

* Static Data Structure (fixed memory)
    * Array
* Dynamic Data Structure (dynamic memory)
    * Queue
    * Stack
    * Linked List

Array — store a collection of elements of the same data type. Elements are allocated contiguous memory, allowing for constant-time access. Each element has a unique index number.

Matrix/Grid — a two-dimensional array of elements, arranged in rows and columns. Represented as a rectangular grid, with each element at the intersection of a row and column.

String — sequence of characters, typically used to represent text.

Stack — follows the Last In, First Out (LIFO) principle. Plays an important role in managing function calls, memory, and is widely used in algorithms, web development, and systems like compilers and browsers

Queue — follows the First In, First Out (FIFO) principle. Plays an important role in managing tasks or data in order, scheduling, and message handling systems.

Linked List — stores data in nodes, which are connected by pointers. Not stored in contiguous memory locations and can only be accessed sequentially, starting from the head of the list.

### Non-Linear Data Structure


Data elements are not placed sequentially or linearly. All elements cannot be traversed on a single run only.

* Tree
* Graph

Tree — hierarchical data structure consisting of nodes connected by edges, with a top node called the root and nodes having child nodes. Widely used in file systems, databases, and decision-making algorithms.

Binary Tree — hierarchical data structure where each node has at most two children, referred to as the left child and right child.

Graph — consists of a finite set of vertices(nodes) and a set of edges(links) that connect a pair of nodes. Widely used to represent relationships between entities.


### Abstract Data Types


Data Structure (DS) — a way of organizing data so that it can be used efficiently. 
Abstract Data Type (ADT) — an abstraction of a data structure which provides only the interface to which a data structure must adhere to.





## Computational Complexity


Big-O Notation
Provides an upper bound of the complexity in the worst case, helping to quantify performance as the input size becomes arbitrarily large.

Time Complexity	Big O Notation
Constant Time	O(1)
Logarithmic Time	O(log(n))
Linear Time	O(n)
Linearithmic Time	0(nlog(n))
Quadratic Time	O(n^2)
Cubic Time	O(n^3)
Exponential Time	O(b^n), b >1
Factorial Time	O(n!)

Complexities, from smallest to largest
n - The size of the input
