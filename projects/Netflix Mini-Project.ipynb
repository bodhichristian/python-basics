```python
import pandas as pd
office_df = pd.read_csv('the_office_series.csv')
office_df.head()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()

colors = []
sizes = []

for rating in office_df['ratings']:
    if rating < 7:
        colors.append('red')
    elif rating < 8:
        colors.append('orange')
    elif rating < 9:
        colors.append('lightgreen')
    else:
        colors.append('darkgreen')
        
for value in office_df['has_guests']:
    size = 25 if value == False else 250
    sizes.append(size)
    
office_df['colors'] = colors
office_df['sizes'] = sizes

x = office_df['episode_number']
y = office_df['viewership_mil']
plt.title('Popularity, Quality, and Guest Appearances on the Office')
plt.xlabel('Episode Number')
plt.ylabel('Viewership (Millions)')
plt.scatter(x, y, s=sizes, c=colors)
plt.show()
```
