# Datetime

```python
import datetime

# dir(datetime)
# help(datetime.date)
```

```python
gvr_birthday = datetime.date(1956, 1, 31)
print(gvr_birthday)
print(gvr_birthday.year)
print(gvr_birthday.month)
print(gvr_birthday.day)
```

## Time Delta

```python
y2k = datetime.date(2000, 1, 1)
delta = datetime.timedelta(100) # takes number of days as arg
print(y2k + delta)
```

## Date formatting

```python
# Day name, Month name, Day #, Year
print(gvr_birthday.strftime('%A, %B, %d, %Y'))
```

# Modern approach

```python
message = 'GVR was born on {:%A, %B, %d, %Y}'
print(message.format(gvr_birthday))
```

## Date, Time, Datetime

```python
launch_date = datetime.date(2017, 3, 30)
launch_time = datetime.time(22, 27, 0)
launch_datetime = datetime.datetime(2017, 3, 30, 22, 27, 0)

print(launch_date)
print(launch_time)
print(launch_datetime)
```

## Access components

```python
print(launch_time.hour)
print(launch_time.minute)
print(launch_time.second)
```

```python
print(launch_datetime.year)
print(launch_datetime.month)
print(launch_datetime.day)
print(launch_datetime.hour)
print(launch_datetime.minute)
print(launch_datetime.second)
```

## Access current datetime

```python
now = datetime.datetime.today()
print(now)
```

```python
print(now.microsecond)
```

## Convert string to datetime

```python
moon_landing = '7/20/1969'
moon_landing_dt = datetime.datetime.strptime(moon_landing, '%m/%d/%Y')
print(moon_landing_dt)
print(type(moon_landing_dt))
```
